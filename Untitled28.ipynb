```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_squared_error
np.random.seed(42)

# range of x
N = 100
x = np.linspace(0, 10, N)
epsilon = np.random.normal(0, 1, N)  # 误差项

#  y = 2x² + 3x + epsilon
y = 2 * x**2 + 3 * x + epsilon
#transfer to 2 D
x = x.reshape(-1, 1)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# linear model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear)

# Polynomial（degree=2）
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)
mse_poly = mean_squared_error(y_test, y_pred_poly)

print(f"Linear Regression MSE: {mse_linear:.4f}")
print(f"Polynomial Regression (degree=2) MSE: {mse_poly:.4f}")
```

```python
# standilazation
scaler = StandardScaler()
X_train_poly_scaled = scaler.fit_transform(X_train_poly)
X_test_poly_scaled = scaler.transform(X_test_poly)

# possible alpha values
alpha_values = [0.001, 0.01, 0.1, 1, 10, 100]

# Ridge
best_ridge_alpha = min(alpha_values, key=lambda alpha:
                        -np.mean(cross_val_score(Ridge(alpha=alpha), X_train_poly_scaled, y_train, cv=5, scoring='neg_mean_squared_error')))
ridge_model = Ridge(alpha=best_ridge_alpha)
ridge_model.fit(X_train_poly_scaled, y_train)
y_pred_ridge = ridge_model.predict(X_test_poly_scaled)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)

# Lasso
best_lasso_alpha = min(alpha_values, key=lambda alpha:
                        -np.mean(cross_val_score(Lasso(alpha=alpha, max_iter=10000), X_train_poly_scaled, y_train, cv=5, scoring='neg_mean_squared_error')))
lasso_model = Lasso(alpha=best_lasso_alpha, max_iter=10000)
lasso_model.fit(X_train_poly_scaled, y_train)
y_pred_lasso = lasso_model.predict(X_test_poly_scaled)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)

print(f"Best Ridge α: {best_ridge_alpha}, MSE: {mse_ridge:.4f}")
print(f"Best Lasso α: {best_lasso_alpha}, MSE: {mse_lasso:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt


sort_idx = np.argsort(X_test.ravel())  # 获取排序索引
X_test_sorted = X_test[sort_idx]  # 重新排序 X_test
y_pred_linear_sorted = y_pred_linear[sort_idx]
y_pred_poly_sorted = y_pred_poly[sort_idx]
y_pred_ridge_sorted = y_pred_ridge[sort_idx]
y_pred_lasso_sorted = y_pred_lasso[sort_idx]

plt.figure(figsize=(10, 6))

# label value
plt.scatter(x, y, color='gray', alpha=0.5, label="True Data")

# linear regression
plt.plot(X_test_sorted, y_pred_linear_sorted, label="Linear Regression", linestyle="dashed", color="blue", linewidth=2)

# polynomial
plt.plot(X_test_sorted, y_pred_poly_sorted, label="Polynomial Regression (degree=2)", linestyle="dashed", color="green", linewidth=2)

# Ridge regression
plt.plot(X_test_sorted, y_pred_ridge_sorted, label=f"Ridge Regression (α={best_ridge_alpha})", linestyle="dashed", color="red", linewidth=2)

# Lasso regression
plt.plot(X_test_sorted, y_pred_lasso_sorted, label=f"Lasso Regression (α={best_lasso_alpha})", linestyle="dashed", color="purple", linewidth=2)

plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Comparison of Regression Models")
plt.show()
```
