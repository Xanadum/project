```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error


np.random.seed(42)
N = 100
x = np.linspace(0, 10, N)
epsilon = np.random.normal(0, 1, N)
y = 3 * x + epsilon  # label function.

# transfer to two D
x = x.reshape(-1, 1)
```

```python
# 80% training set, 20% testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# creat linear regression model
model = LinearRegression()
model.fit(x_train, y_train)

# predict test set
y_pred = model.predict(x_test)

# test mse error
mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE: {mse:.4f}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_train, y_train, color='blue', label='Training data')
plt.scatter(x_test, y_test, color='red', label='Test data')
plt.plot(x, model.predict(x), color='black', linewidth=2, label='Regression Line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression on Simulated Data")
plt.legend()
plt.show()
```

```python
# define different sigma
noise_levels = [0.1, 0.5, 2.0]
mse_values = []

plt.figure(figsize=(12, 4))

for i, sigma in enumerate(noise_levels): # use for to repeat this process
    np.random.seed(42)
    epsilon_new = np.random.normal(0, sigma, N)
    y_new = 3 * x.flatten() + epsilon_new

    # divide training set and testing set
    x_train, x_test, y_train, y_test = train_test_split(x, y_new, test_size=0.2, random_state=42)

    # training set
    model = LinearRegression()
    model.fit(x_train, y_train)

    #  MSE
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    mse_values.append(mse)

    # plot picture
    plt.subplot(1, 3, i+1)
    plt.scatter(x_train, y_train, color='blue', label='Training data')
    plt.scatter(x_test, y_test, color='red', label='Test data')
    plt.plot(x, model.predict(x), color='black', linewidth=2, label='Regression Line')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title(f"Noise  = {sigma}, MSE = {mse:.4f}")
    plt.legend()

plt.tight_layout()
plt.show()

# output different MSE
for sigma, mse in zip(noise_levels, mse_values):
    print(f"Noise  = {sigma}: Test MSE = {mse:.4f}")
```
